## KNN 交叉验证-算法调优

```python
## 加载数据 拆分\
import numpy as np
from sklearn import datasets  #示例数据集
from sklearn.neighbors import KNeighborsClassifier  #分类
```

```python
#X 表示数据，y目标值
X,y = datasets.load_iris(return_X_y=True)
display(X,y)
```

```python
# train 训练数据 test 测试数据
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

X_train,X_test,y_train,y_test=train_test_split(X,y,#要拆分数据
                 test_size=0.2,#拆分比例，测试数据20%
                 random_state=42)#固定了随机状态
display(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
#### 超参数-交叉验证筛选
params_k = np.arange(1,31)

scores = []
for k in params_k:
    knn=KNeighborsClassifier(n_neighbors=k)
    score=cross_val_score(knn,X_train,y_train,cv=5,scoring='accuracy').mean()
    scores.append(score)
display(scores)
```

```python
k_best=np.argmax(scores)+1
print('得分最高的超参数K值是',k_best)
```

#### 选择最佳超参数建模和预测

```python
estmator = KNeighborsClassifier(n_neighbors=k_best)
estmator.fit(X_train,y_train)
```

```python
y_=estmator.predict(X_test)
```

```python
print(y_)
```

```python
print(y_test)
```

```python
estmator.score(X_test,y_test)
```
